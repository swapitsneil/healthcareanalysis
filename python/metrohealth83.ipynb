```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("MetroHealth83.csv")
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.columns = [
    col.strip().lower().replace(" ", "_")
    for col in df.columns
]

# Drop unnecessary index column
if "unnamed:_0" in df.columns:
    df.drop(columns=["unnamed:_0"], inplace=True)

# Converting into numeric columns
numeric_columns = df.columns.drop("city")

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Handle missing values
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
df.shape
```

```python
# New KPI Metrics

df["physician_hospital_ratio"] = df["nummds"] / df["numhospitals"]

df["medicare_growth_flag"] = np.where(
    df["pctchangemedicare"] > 5,
    "High Growth",
    "Normal Growth"
)

# Risk segmentation based on SSI rate
df["risk_category"] = pd.cut(
    df["ssirate"],
    bins=[0, 1500, 3000, np.inf],
    labels=["Low Risk", "Moderate Risk", "High Risk"]
)
```

```python
# Check summary statistics
df.describe()
```

```python
# Data integrity checks
(df[numeric_columns] < 0).sum()
```

```python
# cleaned dataset download

df.to_csv("cleaned_healthcare_data.csv", index=False)

print("\nETL Process Completed Successfully.")
print("Cleaned dataset saved as cleaned_healthcare_data.csv")
```
